### 2. **\$L^2\$ èŒƒæ•° (Euclidean distance, æ¬§å‡ é‡Œå¾—è·ç¦»)**

å…¬å¼ï¼š

### 3. **\$L^\infty\$ èŒƒæ•° (Maximum norm, æœ€å¤§èŒƒæ•°)**

å…¬å¼ï¼š

$$
\|x\|_\infty = \max_i |x_i|
$$

* **ç›´è§‚æ„ä¹‰**ï¼šä»åŸç‚¹åˆ°ç‚¹çš„è·ç¦»ï¼Œç”¨â€œæœ€å¤§åˆ†é‡â€æ¥è¡¡é‡ã€‚æ¢å¥è¯è¯´ï¼Œçœ‹ä½ åœ¨æ‰€æœ‰åæ ‡æ–¹å‘ä¸Šå“ªä¸€ä¸ªæœ€è¿œã€‚

**åº”ç”¨åœºæ™¯**ï¼šå¸¸ç”¨äºè¯¯å·®ä¸Šç•Œçš„åˆ†æã€‚æ¯”å¦‚åœ¨æ•°å€¼è®¡ç®—ä¸­ï¼Œ\$L^\infty\$ èŒƒæ•°è¡¡é‡â€œæœ€åæƒ…å†µä¸‹çš„è¯¯å·®â€ã€‚

### ğŸ” å¯¹æ¯”æ€»ç»“

* \$L^1\$ â†’ èµ°æ ¼å­è·ç¦» â†’ å¼ºè°ƒâ€œæ•´ä½“åç§»é‡â€ã€‚
* \$L^2\$ â†’ ç›´çº¿è·ç¦» â†’ å¼ºè°ƒâ€œå‡ ä½•ä¸Šçš„çœŸå®è·ç¦»â€ã€‚
* \$L^\infty\$ â†’ æœ€å¤§åæ ‡å·® â†’ å¼ºè°ƒâ€œæœ€åæƒ…å†µâ€ã€‚

```python
from numpy.linalg import norm


# L1 norm (Manhattan)
l1_norm = norm(v, 1)
# L2 norm (Euclidean)
l2_norm = norm(v, 2)
# Infinity norm (max-abs)
inf_norm = norm(v, np.inf)


print("L1 norm:", l1_norm)
print("L2 norm:", l2_norm)
print("Infinity norm:", inf_norm)
```

## 3. **Matrix Norms**

å’Œå‘é‡èŒƒæ•°ä¸€æ ·ï¼Œ**çŸ©é˜µèŒƒæ•°**æ˜¯è¡¡é‡çŸ©é˜µâ€œå¤§å°â€çš„å·¥å…·ï¼Œä½†å®ƒå¼ºè°ƒçš„æ˜¯çŸ©é˜µä½œä¸ºçº¿æ€§å˜æ¢çš„ä½œç”¨å¼ºåº¦æˆ–å…ƒç´ æ•´ä½“è§„æ¨¡ã€‚

* **å‡ ä½•è§£é‡Š**ï¼šè¡¡é‡çŸ©é˜µçš„â€œèƒ½é‡â€æˆ–æ€»ä¿¡æ¯é‡ã€‚

#### 2. **Spectral Norm**

#### 3. **Spectral Radius**

$$
\rho(A) = \max \{|\lambda| : \lambda \in \sigma(A)\}
$$

* **ç›´è§‚æ„ä¹‰**ï¼šæ‰€æœ‰ç‰¹å¾å€¼çš„æ¨¡ä¸­å–æœ€å¤§å€¼ï¼Œè¡¨ç¤ºçŸ©é˜µåœ¨æŸä¸ªç‰¹å¾æ–¹å‘ä¸Šçš„æœ€å¤§ä¼¸ç¼©å› å­ã€‚
* **å‡ ä½•è§£é‡Š**ï¼šåˆ»ç”»äº†çŸ©é˜µåœ¨é•¿æœŸä½œç”¨ä¸‹çš„â€œæœ€å¼ºå¢é•¿æ¨¡å¼â€ã€‚
* **åº”ç”¨**ï¼š

  * **è¿­ä»£æ³•æ”¶æ•›æ€§**ï¼šè‹¥è¿­ä»£çŸ©é˜µçš„è°±åŠå¾„ \$\rho(A)<1\$ï¼Œè¿­ä»£è¿‡ç¨‹æ”¶æ•›ã€‚
  * **åŠ¨åŠ›ç³»ç»Ÿç¨³å®šæ€§**ï¼šåœ¨ç¦»æ•£ç³»ç»Ÿ \$x\_{k+1}=Ax\_k\$ ä¸­ï¼Œè‹¥ \$\rho(A)<1\$ï¼Œè§£ä¼šè¶‹å‘äºé›¶ã€‚

```python
# Frobenius norm
fro_norm = norm(M, 'fro')
# Spectral norm (largest singular value)
spectral_norm = norm(M, 2)

print("Frobenius norm:", fro_norm)
print("Spectral norm:", spectral_norm)

# è®¡ç®—è°±åŠå¾„
eigenvalues = np.linalg.eigvals(M)
spectral_radius = max(abs(eigenvalues))

print("Spectral radius:", spectral_radius)
```

* **çŸ©é˜µçš„è§’è‰²**ï¼šçŸ©é˜µä¸ä»…ä»…æ˜¯æ•°å­—è¡¨æ ¼ï¼Œå®ƒæœ¬è´¨ä¸Šæ˜¯ **çº¿æ€§å˜æ¢çš„ç®—å­**ã€‚
* **ä½œç”¨æ–¹å¼**ï¼šå½“çŸ©é˜µ $A$ ä½œç”¨åœ¨å‘é‡ $x$ ä¸Šæ—¶ï¼Œå¾—åˆ°çš„ç»“æœ $Ax$ è¡¨ç¤ºâ€œæŠŠå‘é‡ $x$â€åœ¨ç©ºé—´ä¸­ **æ‹‰ä¼¸ã€ç¼©æ”¾ã€æ—‹è½¬æˆ–åå°„**ã€‚
* **å‡ ä½•æ„ä¹‰**ï¼šçŸ©é˜µå¯ä»¥æŠŠå•ä½æ­£æ–¹å½¢/å•ä½åœ†å˜æ¢æˆå¹³è¡Œå››è¾¹å½¢/æ¤­åœ†ã€‚

---

### 2. **Basis Changeï¼ˆåŸºå˜æ¢ï¼‰**

* **åŸºçš„æ„ä¹‰**ï¼šåŸºï¼ˆbasisï¼‰æ˜¯æè¿°ç©ºé—´ä¸­å‘é‡çš„ä¸€ç»„â€œåæ ‡å°ºâ€ã€‚
* **åŸºå˜æ¢è¿‡ç¨‹**ï¼šåŒä¸€ä¸ªå‘é‡ï¼Œå¯ä»¥åœ¨ä¸åŒåŸºä¸‹æœ‰ä¸åŒçš„åæ ‡è¡¨ç¤ºã€‚

  $$
  [x]_{new} = P^{-1} [x]_{old}
  $$

  å…¶ä¸­ $P$ æ˜¯åŸºå˜æ¢çŸ©é˜µã€‚
* **åº”ç”¨**ï¼š

  * åæ ‡ç³»åˆ‡æ¢ï¼ˆå¦‚ä»ç›´è§’åæ ‡ç³»åˆ°æåæ ‡ç³»ï¼‰ã€‚
  * å¯¹è§’åŒ–ã€ä¸»æˆåˆ†åˆ†æï¼ˆPCAï¼‰ï¼Œé€šè¿‡æ¢åŸºæ­ç¤ºæ•°æ®çš„æœ¬è´¨æ–¹å‘ã€‚

---

ğŸ‘‰ **ä¸€å¥è¯æ€»ç»“**ï¼š

* **çŸ©é˜µ**ï¼šæ˜¯çº¿æ€§å˜æ¢çš„å·¥å…·ï¼Œæ§åˆ¶å‘é‡åœ¨ç©ºé—´ä¸­çš„å½¢çŠ¶å˜åŒ–ã€‚
* **åŸºå˜æ¢**ï¼šæ˜¯é‡æ–°é€‰æ‹©åæ ‡ç³»ï¼Œç”¨æ–°çš„â€œå°ºå­â€æ¥æè¿°åŒä¸€ä¸ªå‘é‡æˆ–å˜æ¢ã€‚


```python
# Linear mapping: matrix acting on vector
A = np.array([[2, 0], [0, 3]])
x = np.array([1, 1])
y = A @ x
print("Linear mapping result y:", y)

# Basis change example
P = np.array([[1, 1], [0, 1]])   # Change of basis matrix
x_new_basis = np.linalg.inv(P) @ x
print("Coordinates of x in new basis:", x_new_basis)

```

### å¸¸è§çš„ä¸‰ç±»å˜æ¢

   * ç”¨æ­£äº¤çŸ©é˜µï¼ˆæ—‹è½¬çŸ©é˜µï¼‰å®ç°ï¼š

     R(\theta) = \begin{bmatrix}

   * ä¿æŒå‘é‡é•¿åº¦å’Œè§’åº¦ä¸å˜ï¼Œåªæ”¹å˜æ–¹å‘ã€‚

     S = \begin{bmatrix}

   * åœ¨ä¸åŒæ–¹å‘ä¸Šæ‹‰ä¼¸æˆ–å‹ç¼©ã€‚

   * æŠŠæ­£æ–¹å½¢å˜æˆå¹³è¡Œå››è¾¹å½¢ï¼Œè§’åº¦å‘ç”Ÿæ”¹å˜ä½†é¢ç§¯å¯èƒ½ä¿æŒã€‚

## 7. Basis Orthogonalization (Gramâ€“Schmidt)

### 1. **é—®é¢˜èƒŒæ™¯**

* åœ¨è®¸å¤šåœºæ™¯ä¸­ï¼ˆå¦‚æ•°å€¼è®¡ç®—ã€ä¿¡å·å¤„ç†ã€PCAï¼‰ï¼Œæˆ‘ä»¬å¸Œæœ›ä¸€ç»„åŸºåº•å‘é‡ **äº’ç›¸æ­£äº¤**ï¼Œè¿™æ ·è®¡ç®—æ›´ç®€æ´ç¨³å®šã€‚
* ä¸€èˆ¬ç»™å®šçš„ä¸€ç»„å‘é‡å¹¶ä¸æ­£äº¤ï¼Œéœ€è¦ç”¨ **Gramâ€“Schmidt æ­£äº¤åŒ–** æ¥å¤„ç†ã€‚

---

### 2. **Gramâ€“Schmidt è¿‡ç¨‹**

è¾“å…¥ï¼šä¸€ç»„çº¿æ€§æ— å…³å‘é‡ $\{v_1, v_2, ..., v_n\}$ã€‚
è¾“å‡ºï¼šä¸€ç»„æ­£äº¤ï¼ˆæˆ–æ­£äº¤å½’ä¸€ï¼‰åŸº $\{u_1, u_2, ..., u_n\}$ã€‚

æ­¥éª¤ï¼š

1. å–ç¬¬ä¸€ä¸ªå‘é‡ï¼š

   $$
   u_1 = v_1
   $$
2. å¯¹åç»­æ¯ä¸ªå‘é‡ï¼Œé€æ­¥å‡å»å®ƒåœ¨å·²æœ‰æ­£äº¤åŸºä¸Šçš„æŠ•å½±ï¼š

   $$
   u_k = v_k - \sum_{j=1}^{k-1} \frac{\langle v_k, u_j \rangle}{\langle u_j, u_j \rangle} u_j
   $$
3. è‹¥éœ€è¦æ­£äº¤**å½’ä¸€**åŸºï¼Œå†æŠŠæ¯ä¸ª $u_k$ å½’ä¸€åŒ–ï¼š

   $$
   e_k = \frac{u_k}{\|u_k\|}
   $$

### 3. **çº¿æ€§æ— å…³**
ä¸€ç»„ **æ­£äº¤å‘é‡**

$$
u_1 = (1,0), \quad u_2 = (0,1)
$$

* å†…ç§¯ï¼š

  $$
  \langle u_1, u_2 \rangle = 1 \cdot 0 + 0 \cdot 1 = 0
  $$
* âœ… å®ƒä»¬æ­£äº¤ï¼ˆäº’ç›¸å‚ç›´ï¼‰ï¼Œè€Œä¸”æ˜¾ç„¶çº¿æ€§æ— å…³ã€‚

---

ä¸€ç»„ **çº¿æ€§æ— å…³ä½†ä¸æ­£äº¤çš„å‘é‡**

$$
v_1 = (1,0), \quad v_2 = (1,1)
$$

* å†…ç§¯ï¼š

  $$
  \langle v_1, v_2 \rangle = 1 \cdot 1 + 0 \cdot 1 = 1 \neq 0
  $$
* âŒ å®ƒä»¬ä¸æ­£äº¤ï¼ˆä¸å‚ç›´ï¼‰ã€‚
* ä½†å®ƒä»¬ **çº¿æ€§æ— å…³**ï¼Œå› ä¸ºï¼š

  $$
  v_2 \neq \alpha v_1 \quad \text{å¯¹ä»»æ„å®æ•° }\alpha
  $$

  ï¼ˆå³ $v_2$ ä¸æ˜¯ $v_1$ çš„å€æ•°ï¼‰ã€‚

ğŸ‘‰ è¿™è¯´æ˜ï¼šè¿™ä¸¤ä¸ªå‘é‡è™½ç„¶â€œæ­ªæ­ªæ–œæ–œâ€ï¼Œä½†ä¾ç„¶èƒ½å¼ æˆæ•´ä¸ªäºŒç»´å¹³é¢ã€‚

---
ç›´è§‚è§£é‡Š

* â€œçº¿æ€§æ— å…³â€ = å®ƒä»¬æŒ‡å‘ä¸åŒçš„æ–¹å‘ï¼Œèƒ½è¦†ç›–æ›´å¤§çš„ç©ºé—´ã€‚
* â€œæ­£äº¤â€ = ä¸ä»…æ–¹å‘ä¸åŒï¼Œè€Œä¸”å®Œå…¨æ²¡æœ‰é‡å åˆ†é‡ã€‚
* æ‰€ä»¥ï¼š

  * æ‰€æœ‰æ­£äº¤å‘é‡ç»„ä¸€å®šçº¿æ€§æ— å…³ã€‚
  * ä½†å¹¶ä¸æ˜¯æ‰€æœ‰çº¿æ€§æ— å…³å‘é‡ç»„éƒ½æ­£äº¤ã€‚

---

âœ… ä¸€å¥è¯æ€»ç»“ï¼š
$(1,0)$ å’Œ $(1,1)$ æ˜¯ **çº¿æ€§æ— å…³ä½†ä¸æ­£äº¤** çš„å…¸å‹ä¾‹å­ã€‚

---

### 3. **ç›´è§‚è§£é‡Š**

* å°±åƒæŠŠä¸€ç»„â€œæ­ªæ–œâ€çš„å‘é‡ï¼Œé€ä¸ªâ€œæŠ•å½±å¹¶å‡æ‰é‡å éƒ¨åˆ†â€ï¼Œæœ€åå¾—åˆ°ä¸€ç»„äº’ç›¸å‚ç›´çš„â€œæ–°å°ºå­â€ã€‚
* **å‡ ä½•æ„ä¹‰**ï¼šåœ¨åŸç©ºé—´ä¸å˜çš„å‰æä¸‹ï¼Œé‡æ–°é€‰ä¸€ç»„å½¼æ­¤æ­£äº¤çš„åŸºã€‚

---

### 4. **åº”ç”¨**

* QR åˆ†è§£çš„åŸºç¡€ï¼ˆçŸ©é˜µåˆ†è§£ã€æ•°å€¼ç¨³å®šè®¡ç®—ï¼‰ã€‚
* ç®€åŒ–å‡ ä½•æˆ–ä»£æ•°è¿ç®—ï¼ˆæ­£äº¤åŸºä¸‹ç‚¹ç§¯ã€æŠ•å½±æ›´æ–¹ä¾¿ï¼‰ã€‚
* æ•°æ®é™ç»´ï¼ˆPCA ä¸­çš„æ­£äº¤åŒ–æ­¥éª¤ï¼‰ã€‚

---

ğŸ‘‰ **ä¸€å¥è¯æ€»ç»“**ï¼š
Gramâ€“Schmidt æ­£äº¤åŒ–å°±æ˜¯ä¸€ç§ç³»ç»Ÿæ–¹æ³•ï¼ŒæŠŠä»»æ„ä¸€ç»„çº¿æ€§æ— å…³å‘é‡â€œä¿®æ­£â€ä¸ºäº’ç›¸æ­£äº¤ï¼ˆæˆ–æ ‡å‡†æ­£äº¤ï¼‰çš„åŸºåº•ã€‚


```python

def gram_schmidt(vectors):
    orthogonal = []
    for v in vectors:
        w = v - sum(np.dot(v, u)/np.dot(u, u) * u for u in orthogonal)
        if np.linalg.norm(w) > 1e-10:
            orthogonal.append(w)
    return np.array(orthogonal)

# Example basis
B = np.array([[1, 1], [1, 0]], dtype=float)
orthogonal_basis = gram_schmidt(B)

print("Original basis:\n", B)
print("Orthogonalized basis:\n", orthogonal_basis)
```
