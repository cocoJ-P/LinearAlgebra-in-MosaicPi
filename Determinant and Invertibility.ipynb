# 2. Determinant and Invertibility

The **determinant** of a square matrix $A$ encodes both algebraic and geometric properties:

---

1. **Determinant as a Volume Scaling Factor**
- $\det(A)$ measures how the linear transformation defined by $A$ scales volume.
- In 2D: absolute value of $\det(A)$ gives the area scaling factor of a unit square.
- In 3D: absolute value of $\det(A)$ gives the volume scaling factor of a unit cube.
- If $\det(A) < 0$, the transformation includes a reflection (orientation change).

---

2. **Invertibility Criterion**
- A matrix $A$ is **invertible** (non-singular) if and only if $\det(A) \neq 0$.
- If $\det(A) = 0$, the transformation **collapses** space into a lower dimension:
  - Lines â†’ points
  - Surfaces â†’ lines
- This loss of dimension means no inverse mapping exists.

---

3. **Jacobian and Local Invertibility**
For an implicit or explicit function $y = f(x)$:

- The **Jacobian matrix** $J = \frac{\partial{f}}{\partial{x}}$ describes local linearization.
- $\det(J)$ measures the **local volume distortion** of the mapping.
- **Cases:**
  - $\det(J) \neq 0$ â†’ locally invertible mapping (implicit function theorem applies).
  - $\det(J) = 0$ â†’ **singularity or bifurcation point**, where:
    - The mapping may lose uniqueness.
    - Critical transitions (e.g., fold, cusp bifurcations) can occur.

---

**Summary:**  
The determinant is not just a formula â€” it provides a **bridge** between algebra (invertibility) and geometry (volume/orientation). In applied contexts like nonlinear systems, checking $\det(J)$ is a fundamental step for analyzing **stability, bifurcations, and solvability**.

# 3. Jacobian and Local Invertibility â€” Details

## 3.1 å±€éƒ¨çº¿æ€§åŒ–ï¼ˆJacobian çš„è§’è‰²ï¼‰

è®¾ $f:\mathbb R^n\to\mathbb R^n$ å¯å¾®ï¼Œåœ¨ç‚¹ $x_0$ å¤„çš„ä¸€é˜¶æ³°å‹’å±•å¼€ï¼š

$$

$$

* **å«ä¹‰**ï¼šåœ¨è¶³å¤Ÿå°çš„é‚»åŸŸé‡Œï¼Œ$f$ å°±åƒâ€œå…ˆåšä¸€ä¸ªçº¿æ€§å˜æ¢ $J(x_0)$â€å†å¹³ç§»ã€‚
* **å¥‡å¼‚å€¼è§†è§’**ï¼šè‹¥æŠŠ $J$ åš SVDï¼Œå¥‡å¼‚å€¼ $\sigma_i$ ç»™å‡ºåœ¨æ­£äº¤æ–¹å‘ä¸Šçš„**å±€éƒ¨æ‹‰ä¼¸å€æ•°**ï¼›$|\det J|=\prod_i \sigma_i$ æ˜¯**å±€éƒ¨ä½“ç§¯ç¼©æ”¾**ã€‚

## 3.2 $\det(J)$ çš„å‡ ä½•æ„ä¹‰ï¼šå±€éƒ¨ä½“ç§¯ä¸æ–¹å‘

* $|\det J(x_0)|$ï¼šæŠŠä¸€ä¸ª**æ— é™å°ä½“ç§¯**å…ƒç´ æ˜ å°„åçš„ä½“ç§¯å€æ•°ï¼ˆå±€éƒ¨ä½“ç§¯ç¼©æ”¾å› å­ï¼‰ã€‚
* $\det J>0$ï¼šä¿æŒæ–¹å‘ï¼ˆorientationï¼‰ï¼›$\det J<0$ï¼šåŒ…å«ä¸€æ¬¡åå°„ï¼ˆç¿»è½¬æ–¹å‘ï¼‰ã€‚

> åªæœ‰å½“ $n=n$ï¼ˆæ–¹é˜µ Jacobianï¼‰æ—¶æ‰è°ˆ $\det$ã€‚æ›´ä¸€èˆ¬åœ°ç”¨â€œç§©/å¥‡å¼‚å€¼â€è¡¨è¿°å±€éƒ¨è¡Œä¸ºï¼š
>
> * **æ»¡ç§©**ï¼ˆæ‰€æœ‰ $\sigma_i>0$ï¼‰ï¼šæ— ç»´åº¦åç¼©ï¼›
> * **ç§©äº**ï¼ˆæŸäº› $\sigma_i=0$ï¼‰ï¼šå±€éƒ¨åç¼©åˆ°ä½ç»´ã€‚

## 3.3 å±€éƒ¨å¯é€†æ€§çš„åˆ¤æ®ï¼ˆåå‡½æ•°å®šç†ï¼‰

* **åå‡½æ•°å®šç†**ï¼ˆæ˜¾å¼æ˜ å°„ $y=f(x)$ï¼‰ï¼šè‹¥ $J(x_0)$ **å¯é€†**ï¼ˆ$\det J(x_0)\neq 0$ï¼‰ï¼Œåˆ™ $f$ åœ¨ $x_0$ é™„è¿‘æ˜¯**å±€éƒ¨åŒå°„**ä¸”å¯å¾®ï¼Œå­˜åœ¨å…‰æ»‘çš„å±€éƒ¨é€† $f^{-1}$ã€‚
* **éšå‡½æ•°å®šç†**ï¼ˆçº¦æŸ $F(x,y)=0$ï¼‰ï¼šè®¾ $F:\mathbb R^{m+n}\to\mathbb R^n$ï¼Œè‹¥

  $$
  F(x_0,y_0)=0,\quad \det\Big(\frac{\partial F}{\partial y}(x_0,y_0)\Big)\neq 0,
  $$

  åˆ™å­˜åœ¨é‚»åŸŸä¸**å”¯ä¸€**å…‰æ»‘å‡½æ•° $y=g(x)$ ä½¿ $F(x,g(x))=0$ã€‚

> ç›´è§‰ï¼šJacobian å¯é€† $\Leftrightarrow$ æ²¡æœ‰æ–¹å‘è¢«å‹æ‰æ‰ $\Leftrightarrow$ å±€éƒ¨ä¸€ä¸€å¯¹åº”ã€‚

## 3.4 $\det(J)=0$ï¼šå¥‡å¼‚ç‚¹ä¸åˆ†å²”

å½“ $\det J=0$ æ—¶ï¼Œè‡³å°‘æœ‰ä¸€ä¸ªæ–¹å‘è¢«å‹æ‰ï¼ˆæŸä¸ªå¥‡å¼‚å€¼ä¸º 0ï¼‰ï¼Œå…¸å‹åæœï¼š

* **å¤±å»å”¯ä¸€æ€§**ï¼šå¤šä¸ªè¾“å…¥æ˜ å°„åˆ°åŒä¸€è¾“å‡ºï¼Œæˆ–è€…åŒä¸€çº¦æŸä¸‹å‡ºç°å¤šè§£ã€‚
* **ä¸´ç•Œè½¬å˜/åˆ†å²”**ï¼šå‚æ•°å˜åŒ–åˆ°æŸä¸ªä¸´ç•Œå€¼æ—¶ï¼Œè§£çš„ç»“æ„å‘ç”Ÿå®šæ€§æ”¹å˜ï¼ˆå¦‚åˆå¹¶/è¯ç”Ÿ/äº¤æ¢ç¨³å®šæ€§ï¼‰ã€‚

### ç»å…¸å°ä¾‹å­

**(A) æ˜¾å¼æ˜ å°„ï¼šå¤æ‚å¹³æ–¹æ˜ å°„ï¼ˆäºŒç»´ï¼‰**

$$

$$

Jacobianï¼š

$$
J=\begin{bmatrix}2x&-2y\\ 2y&2x\end{bmatrix},\quad \det J=4(x^2+y^2).
$$

* é™¤åŸç‚¹å¤– $\det J>0$ â‡’ å±€éƒ¨å¯é€†ã€‚
* åœ¨åŸç‚¹ $\det J=0$ â‡’ **åˆ†æ”¯ç‚¹**ï¼šå…¨å±€ä¸Šè¯¥æ˜ å°„æ˜¯â€œä¸¤å¯¹ä¸€â€ï¼Œåœ¨åŸç‚¹å¤„åˆå¹¶ã€‚

**(B) éšå‡½æ•°ï¼š$F(x,y)=y^2-x=0$**

$$
\frac{\partial F}{\partial y}=2y.

* åœ¨ $(x_0,y_0)$ ä¸” $y_0\neq 0$ å¤„ï¼Œ$\partial F/\partial y\neq 0$ â‡’ å­˜åœ¨å”¯ä¸€æ˜¾å‡½æ•° $y=\pm\sqrt{x}$ çš„å…¶ä¸­ä¸€æ”¯ï¼ˆå–å†³äºç‚¹æ‰€åœ¨åˆ†æ”¯ï¼‰ã€‚
* åœ¨ $(0,0)$ å¤„ï¼Œ$\partial F/\partial y=0$ â‡’ **ä¸å”¯ä¸€**ï¼šåŒä¸€ç‚¹ $x$ çš„é‚»åŸŸå¯¹åº”ä¸¤æ¡åˆ†æ”¯ï¼ˆä¸Š/ä¸‹ï¼‰ï¼Œè¿™æ˜¯\*\*æŠ˜å ï¼ˆfoldï¼‰\*\*å‹å¥‡å¼‚ç‚¹ã€‚

**(C) åŠ¨åŠ›ç³»ç»Ÿçš„åˆ†å²”ï¼ˆå‚æ•° $\mu$ï¼‰**
å¹³è¡¡ç‚¹ç”± $F(x,\mu)=0$ å®šä¹‰ã€‚ä»¥

$$

$$

ä¸ºä¾‹ï¼ˆ**éç»“/æŠ˜å åˆ†å²”**ï¼‰ï¼š

* $\partial F/\partial x=-2x$ã€‚
* å½“ $\mu>0$ æœ‰ä¸¤è§£ $x=\pm\sqrt{\mu}$ï¼›$\mu<0$ æ— è§£ï¼›$\mu=0$ æ—¶åœ¨ $(x,\mu)=(0,0)$ å¤„ $\partial F/\partial x=0$ â€”â€” **Jacobian å¥‡å¼‚**ï¼Œä¸¤æ”¯åœ¨æ­¤**ç”Ÿæˆ/æ¶ˆå¤±**ã€‚

## 3.5 å®æ“ä¸æ•°å€¼è§†è§’

* **è®¡ç®—åˆ¤æ®**ï¼š$\det J\neq 0$ â‡’ å±€éƒ¨å¯é€†ï¼›$\det J=0$ â‡’ å¯èƒ½å¤šè§£/å¥‡å¼‚/ä¸å¯é€†ã€‚
* **æ•°å€¼ç¨³å¥æ€§**ï¼š$|\det J|$ å¾ˆå°ä½†éé›¶ â‡’ è¿‘ä¼¼å¥‡å¼‚ï¼Œå±€éƒ¨é€†è™½å­˜åœ¨ä½†**ç—…æ€**ï¼ˆå¯¹æ‰°åŠ¨éå¸¸æ•æ„Ÿï¼‰ã€‚

  * æ›´ç¨³å¦¥ï¼šçœ‹ $J$ çš„**æœ€å°å¥‡å¼‚å€¼** $\sigma_{\min}$ å’Œæ¡ä»¶æ•° $\kappa(J)=\sigma_{\max}/\sigma_{\min}$ã€‚

## 3.6 ä¸€é¡µå°ç»“

* **Jacobian** æä¾›å‡½æ•°åœ¨æŸç‚¹çš„ **æœ€ä½³çº¿æ€§è¿‘ä¼¼**ã€‚
* $|\det J|$ = å±€éƒ¨ä½“ç§¯ç¼©æ”¾ï¼›ç¬¦å· = æ˜¯å¦ç¿»è½¬æ–¹å‘ã€‚
* **$\det J\neq 0$**ï¼šå±€éƒ¨å¯é€†ï¼ˆåå‡½æ•°å®šç†/éšå‡½æ•°å®šç†é€‚ç”¨ï¼‰ã€‚
* **$\det J=0$**ï¼šå±€éƒ¨å¥‡å¼‚/åˆ†å²”çš„å¯èƒ½ç‚¹ï¼Œå¯èƒ½å¤±å»å”¯ä¸€æ€§æˆ–å‘ç”Ÿè§£ç»“æ„è½¬å˜ã€‚

```python
import sympy as sp

# Example: nonlinear mapping f(x,y) = (x^2, y^2)
x, y = sp.symbols('x y')
f1 = x**2
f2 = y**2
J = sp.Matrix([f1, f2]).jacobian([x, y])
print("Jacobian matrix:\n", J)

det_J = J.det()
print("det(J):", det_J)

# Evaluate at a point
det_at = det_J.subs({x:1, y:2})
print("det(J) at (1,2):", det_at)
```
